```javascript
const express = require('express')
const app = express()
const port = 3000

app.get('/', (req, res) => {
  res.send('Hello World!')
})

app.listen(port, () => {
  console.log(`Example app listening on port ${port}`)
})

```

```javascript
var child_process = require('child_process');
//child_process.exec('npm install -g ffi',{stdio:[0,1,2]});

child_process.exec(`npm --version`, (err, stdout, stderr) => { 
    console.log("npm version:",stdout) // 8.1.2  (in the end of the output)
 });

child_process.exec(`node --version`, (err, stdout, stderr) => { 
    console.log("node version:",stdout) // v16.13.1  (in the end of the output)
 });
```
